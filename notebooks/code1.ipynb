```python
# IMPORTAÇÃO DE BIBLIOTECAS
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# CRIAÇÃO DE UM DATAFRAME COM CSV

df = pd.read_csv("../data/globalterrorism.csv", encoding='latin1')
```

```python
# VAMOS VER AS PRIMEIRAS 5 LINHAS DO DATAFRAME

df.head()
```

```python
# AGRUPAR O NÚMERO DE ATAQUES POR ANO E FAZER UMA NOVO CSV COM ESSAS INFORMAÇÕES, NESSE CASO APENAS 48 LINHAS 
df_decade = df[['iyear', 'eventid']]
attacks_per_year = df_decade.groupby('iyear').count().reset_index()
attacks_per_year.columns = ['Year', 'Number of Attacks']
attacks_per_year.to_csv("../data/attacks_per_year.csv", index=False)
```

```python
# AGRUPAR O NÚMERO DE ATAQUES POR REGIÃO E FAZER UMA NOVO CSV COM ESSAS INFORMAÇÕES, NESSE CASO APENAS 13 LINHAS
df_region = df[['region_txt', 'eventid']]  
attacks_per_region = df_region.groupby('region_txt').count().reset_index()
attacks_per_region.columns = ['Region', 'Number of Attacks']
attacks_per_region.to_csv("../data/attacks_per_region.csv", index=False)
```

```python
# CRIAR DOIS CSVS, UM PARA ATAQUES POR REGIÃO NA DECADA DE 1970 E OUTRO PARA A DE 2010
df_1970s = df[(df['iyear'] >= 1970) & (df['iyear'] < 1980)]
df_2010s = df[(df['iyear'] >= 2010) & (df['iyear'] < 2020)]
attacks_1970s = df_1970s[['region_txt', 'eventid']].groupby('region_txt').count().reset_index()
attacks_1970s.columns = ['Region', 'Number of Attacks']
attacks_1970s.to_csv("../data/attacks_per_region_1970s.csv", index=False)
attacks_2010s = df_2010s[['region_txt', 'eventid']].groupby('region_txt').count().reset_index()
attacks_2010s.columns = ['Region', 'Number of Attacks']
attacks_2010s.to_csv("../data/attacks_per_region_2010s.csv", index=False) 
```

```python
# CRIAR DOIS CSVS, UM PARA TIPOS DE ATAQUES MAIS FREQUENTES E OUTRO COM OS PRINCIPAIS ALVOS DOS ATAQUES
top_attack_types = df['attacktype1_txt'].value_counts().reset_index()
top_attack_types.columns = ['Attack Type', 'Number of Attacks']
top_attack_types.to_csv("../data/top_attack_types.csv", index=False)
top_target_types = df['targtype1_txt'].value_counts().reset_index()
top_target_types.columns = ['Target Type', 'Number of Attacks']
top_target_types.to_csv("../data/top_target_types.csv", index=False)
```
