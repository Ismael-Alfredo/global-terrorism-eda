```python
# IMPORTAÇÃO DAS BIBLIOTECAS

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# CRIAÇÃO DE UM DATAFRAME COM CSV

df = pd.read_csv("../data/globalterrorism.csv", encoding='latin1')
```

```python
# VAMOS VER AS PRIMEIRAS 5 LINHAS DO DATAFRAME

df.head()
```

```python
# COMO O DATASET É GRANDE, VAMOS VER AS ULTIMAS 5 LINHAS TAMBÉM

df.tail()
```

```python
# CONFIGURAÇÕES PARA VER TODAS AS COLUNAS DO DATAFRAME

pd.set_option('display.max_columns', None)
df.head()
```

```python
# EDA 1 - DE 1970 À 2017, O NÚMERO DE ATAQUES TERRORISTAS AUMENTOU OU DIMINUIU?

plt.figure(figsize=(24,6))
sns.countplot(data=df, x='iyear', palette='magma')
```

```python
# EDA 2 - QUAL FOI O ANO COM MAIS ATAQUES TERRORISTAS?

year_counts = df.groupby('iyear').size()
max_year = year_counts.idxmax()
print(f"O ano com mais ataques terroristas foi {max_year} com {year_counts[max_year]} ataques.")
```

```python
# EDA 3 - QUAL FOI O ANO COM MENOS ATAQUES TERRORISTAS?

min_year = year_counts.idxmin()
print(f"O ano com menos ataques terroristas foi {min_year} com {year_counts[min_year]} ataques.")
```

```python
# EDA 4 - QUAL REGIÃO DO MUNDO SOFREU MAIS ATAQUES TERRORISTAS ENTRE 1970 E 2017?

region_counts = df['region_txt'].value_counts()
max_region = region_counts.idxmax()
print(f"A região com mais ataques terroristas foi {max_region} com {region_counts[max_region]} ataques.")
```

```python
# EDA 5 - QUAL FOI A REGIÃO DO MUNDO COM MENOS ATAQUES TERRORISTAS ENTRE 1970 E 2017?

min_region = region_counts.idxmin()
print(f"A região com menos ataques terroristas foi {min_region} com {region_counts[min_region]} ataques.")
```

```python
# EDA 6 - QUAIS ERAM AS TENDÊNCIAS DOS ATAQUES NA DÊCADA DE 1970? (REGIÕES E FREQUÊNCIA)

decade_70s = df[(df['iyear'] >= 1970) & (df['iyear'] < 1980)]
decade_70s_region_counts = decade_70s['region_txt'].value_counts()
plt.figure(figsize=(24,6))
sns.barplot(x=decade_70s_region_counts.index, y=decade_70s_region_counts.values, palette='magma')

```

```python
# EDA 7 - QUAIS ERAM AS TENDÊNCIAS DOS ATAQUES NA DÊCADA DE 2010? (REGIÕES E FREQUÊNCIA)

decade_2010s = df[(df['iyear'] >= 2010) & (df['iyear'] < 2020)]
decade_2010s_region_counts = decade_2010s['region_txt'].value_counts()
plt.figure(figsize=(24,6))
sns.barplot(x=decade_2010s_region_counts.index, y=decade_2010s_region_counts.values, palette='magma')  
```

```python
# EDA 8 - QUAL A RELAÇÃO ENTRE A ÉPOCA DO ANO E A FREQUÊNCIA DE ATAQUES TERRORISTAS?

df['imonth'].value_counts().sort_index().plot(kind='bar', figsize=(12,6), color='gray')
```

```python
# EDA 9 - TIPOS DE ATAQUES MAIS FREQUENTES POR REGIÃOI
plt.figure(figsize=(16,16))
sns.countplot(data=df, y='attacktype1_txt', hue='region_txt', order=df['attacktype1_txt'].value_counts().index, palette='magma')
```

```python
# EDA 10 - ALVOS MAIS VISADOS POR REGIÃO

plt.figure(figsize=(16,24))
sns.countplot(data=df, y='targtype1_txt', hue='region_txt', order=df['targtype1_txt'].value_counts().index, palette='magma')
```
